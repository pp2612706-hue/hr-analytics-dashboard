```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

n = 100

departments = ['HR', 'Sales', 'IT', 'Finance', 'Operations']
genders = ['Male', 'Female']
recruitment_sources = ['LinkedIn', 'Referral', 'Job Portal', 'Campus']
attrition_options = ['Yes', 'No']
engagement_levels = ['Low', 'Medium', 'High']
promotion_status = ['Promoted', 'Not Promoted']

df = pd.DataFrame({
    'employee_id': [f'E{str(i).zfill(3)}' for i in range(1, n+1)],
    'department': np.random.choice(departments, n),
    'gender': np.random.choice(genders, n),
    'age': np.random.randint(22, 50, n),
    'experience_years': np.random.randint(1, 20, n),
    'monthly_salary': np.random.randint(25000, 120000, n),
    'satisfaction_score': np.random.randint(3, 10, n),
    'training_hours': np.random.randint(5, 60, n),
    'attendance_percent': np.random.randint(70, 100, n),
    'performance_rating': np.random.randint(1, 6, n),
    'overtime_hours': np.random.randint(0, 25, n),
    'attrition': np.random.choice(attrition_options, n, p=[0.25, 0.75]),
    'recruitment_source': np.random.choice(recruitment_sources, n),
    'engagement_level': np.random.choice(engagement_levels, n),
    'promotion_status': np.random.choice(promotion_status, n)
})

df.head()
```

```python
print("Total Employees:", df.shape[0])
print("Average Satisfaction:", df['satisfaction_score'].mean())
print("Average Attendance:", df['attendance_percent'].mean())
print("Attrition Count:", (df['attrition'] == 'Yes').sum())
print("Average Performance:", df['performance_rating'].mean())
```

```python
df[df['attrition']=='Yes']['department'].value_counts().plot(kind='bar')
plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby('department')['satisfaction_score'].mean().plot(kind='bar')
plt.title("Satisfaction by Department")
plt.show()
```

```python
df.groupby('department')['attendance_percent'].mean().plot(kind='bar')
plt.title("Attendance by Department")
plt.show()
```

```python
df['performance_rating'].plot(kind='hist', bins=5)
plt.title("Performance Distribution")
plt.show()
```

```python
df.groupby('attrition')['overtime_hours'].mean().plot(kind='bar')
plt.title("Overtime vs Attrition")
plt.show()
```

```python
df['recruitment_source'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Recruitment Source")
plt.ylabel('')
plt.show()
```

```python
print("Insights:")
print("High attrition departments need HR action")
print("High overtime may indicate burnout")
print("Low satisfaction shows engagement issue")
```
